Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_circles
```

Generating a dataset

```python
# Create a dataset with two interleaving half circles
X, y = make_circles(n_samples=100, factor=0.5, noise=0.1)
```

```python
X
```

```python
y
```

Training SVM with RBF kernal

```python
# Create an SVM model with an RBF kernel
model = svm.SVC(kernel='rbf', C=1.0, gamma=0.5)

# Fit the model to the data
model.fit(X, y)
```

Plotting Decision Boundaries

```python
# Create a grid to plot the decision boundary
xx, yy = np.meshgrid(np.linspace(-1.5, 1.5, 500), np.linspace(-1.5, 1.5, 500))
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the dataset and decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors='k')
plt.contourf(xx, yy, Z, levels=[Z.min(), 0, Z.max()], alpha=0.3, colors=['#FFAAAA', '#AAAAFF'])
plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM with RBF Kernel')
plt.show()
```
